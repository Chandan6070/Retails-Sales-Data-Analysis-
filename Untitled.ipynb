```python
def RL(x):
    if x == 'I':
        return 1
    elif x == 'V':
        return 2
    elif x == 'X':
        return 10
    elif x == 'L':
        return 50
    elif x == 'C':
        return 100
    elif x == 'D':
        return 500
    return -1

    
def RomToInt(str):
    a =0
    i =0
    while(i<len(str)):
        X1 = RL(str[i])
        if (i+1<len(str)):
            X2 = RL(str[i+1])
            if X1 >= X2:
                a = a+10
                i = i+1
            else:
                a = a +X2
                i = i+1
        else:
            a = a+X1
            i = i + 1

print(RomToInt('V'))
```

```python
def analyze_weather_data(dataframe):
    # Calculate the difference between the maximum and minimum temperature for each city on each day
    dataframe['DailyTempRange'] = dataframe['MaxTemp'] - dataframe['MinTemp']

    # Group the data by city
    grouped_df = dataframe.groupby('City')

    # Calculate the average daily temperature range for each city
    avg_temp_range = grouped_df['DailyTempRange'].mean().reset_index()

    # Find the highest temperature and the corresponding date for each city
    highest_temp = grouped_df['MaxTemp'].idxmax()
    highest_temp_date = dataframe.loc[highest_temp, ['City', 'Date', 'MaxTemp']].reset_index(drop=True)

    # Find the lowest temperature and the corresponding date for each city
    lowest_temp = grouped_df['MinTemp'].idxmin()
    lowest_temp_date = dataframe.loc[lowest_temp, ['City', 'Date', 'MinTemp']].reset_index(drop=True)

    # Merge the dataframes together
    result_df = avg_temp_range.merge(highest_temp_date, on='City').merge(lowest_temp_date, on='City')

    # Rename the columns
    result_df.columns = ['City', 'Average Daily Temp Range', 'Highest Temp Date', 'Highest Temp', 'Lowest Temp Date', 'Lowest Temp']

    # Sort the resulting DataFrame by the average daily temperature range in descending order
    sorted_df = result_df.sort_values(by='Average Daily Temp Range', ascending=False)

    # Print the resulting DataFrame
    print(sorted_df)

analyze_weather_data(df)
```

```python
def analyze_stock_data(dataframe):
    # Calculate the 5-day moving average for each company's stock price
    dataframe['MovingAverage'] = dataframe.groupby('Company')['StockPrice'].rolling(window=5).mean().reset_index(level=0, drop=True)

    # Identify the days on which the stock price of a company was greater than its 5-day moving average
    dataframe['GreaterThanMA'] = dataframe['StockPrice'] > dataframe['MovingAverage']

    # Filter the data
    filtered_df = dataframe[dataframe['GreaterThanMA']]

    # Group the data by the company and calculate the total number of days and average stock price
    
    grouped_df = filtered_df.groupby('Company').agg({'Date': 'count', 'StockPrice': 'mean'}).reset_index()

    # Rename the columns
    grouped_df.columns = ['Company', 'Total Days', 'Average Price']

    # Sort the resulting DataFrame by the total number of days in descending order
    sorted_df = grouped_df.sort_values(by='Total Days', ascending=False)

    # Print the resulting DataFrame
    print(sorted_df)

analyze_stock_data(df)
```
